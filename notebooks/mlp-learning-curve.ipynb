```python
import sys
sys.path.append('..')

import matplotlib.pylab as plt
from tqdm import tqdm
from IPython import display

from src.visualization import plot_learning_curve
from src.experiment_tagging import get_model_path
```

```python
LEARNING_CURVES_PARAMS = [('MNIST', {}),
                          ('FASHION', {}),
                          ('CIFAR10', {}),
                          ('LINE', {}),
                          ('RANDOM', {'secondary_y': ('train_loss', 'test_loss')}),
                          ('MNIST-x10-EPOCHS', {}),
                          ('CIFAR10', {'start': 1, 'step': 1}),
                          ('RANDOM-x50-EPOCHS', {'start': 20, 'step': 20, 'secondary_y': ('train_loss', 'test_loss')}),
                          ('RANDOM-OVERFITTING', {}),
                          ('FASHION', {'start': 1, 'step': 1}),
                          ('MNIST', {'start': 1, 'step': 1}),
                          ('MNIST+DROPOUT', {'start': 1, 'step': 1}),
                          ('FASHION+DROPOUT', {'start': 1, 'step': 1}),
                         ]
```

```python
for model_name, params in tqdm(LEARNING_CURVES_PARAMS):
    print(f'#### {model_name}')
    
    _, ax = plt.subplots(1, figsize=(7, 7))
    ax.set_title(model_name)
    _, df = plot_learning_curve(get_model_path(model_name), ax=ax)

    display.display(df)
```
